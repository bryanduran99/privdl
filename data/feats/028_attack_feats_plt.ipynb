```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import json
```

```python
# load feats
batch_rec = torch.load('/data/feats/028_attack_train_0/rec_feats_0.pth')
batch_perm = torch.load('/data/feats/028_attack_train_0/permuted_feats_0.pth')
batch_raw = torch.load('/data/feats/028_attack_train_0/raw_feats_0.pth')
plt.hist(batch_raw[0][0].view(-1), bins=10)
plt.show()
plt.hist(batch_perm[0][0].view(-1), bins=10)
plt.show()
plt.hist(batch_rec[0][0].view(-1), bins=10)
plt.show()
```

```python
# plot recovered feats
for batch_feats in [batch_raw[0], batch_perm[0], batch_rec[0]]:
    # print(batch_feats.size())
    plt.figure(figsize=(20, 70))
    for index, feats in enumerate(batch_feats):
        if index == 12:
            break
        plt.subplot(1,12,index+1)
        plt.imshow(feats[0, :, :], cmap='summer')
        plt.axis('off')
    plt.show()
```
